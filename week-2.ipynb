### Step:6 - Creating a pairplot and mention the findings

#### Converting String label into Numerical ones.

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_Y=LabelEncoder()

df.iloc[:,0]=labelencoder_Y.fit_transform(df.iloc[:,0].values)
```

I order to plot the graph to analyse relation between variable, I am using pairplot from Seaborn Library. Plotting all variable in one grid makes difficult to analyse the grid and relation, In order to overcome this challenge, I have decided to plot grid for group of variables separately.

#### Plotting Pair plot for "radius_mean", "texture_mean","perimeter_mean","area_mean","smoothness_mean"

```python
import seaborn as snc
snc.pairplot(data=df, vars={"radius_mean", "texture_mean","perimeter_mean","area_mean","smoothness_mean"})
```

```python
df.corr(method="pearson")
```

## Following are the strongly co-related variables.

1. perimeter_mean, radius_mean, area_mean, perimeter_worst, radius_worst, area_worst are mutually strongly correlated.
2. texture_worst and texture_mean
3. concavity_mean and concave points_mean
4. concave points_mean and concave points_worst
5. radius_se, perimeter_se, area_se

From the above analysis, we can reduce or eliminate by using just one variable from strongly co-related variables.
For Example:-

perimeter_mean can be used and radius_mean, area_mean, perimeter_worst, radius_worst, area_worst can be dropped. Similarly, texture_worst or texture_mean can be used and another one can be dropped.


## Conclusion:- 

By analysing the co-relation matrix, we can drop columns such as:- 

radius_mean, area_mean, perimeter_worst, radius_worst, area_worst, texture_mean, concave points_mean, concave points_worst.
Reducing dimensions will improve the performance and reduce the complexity of the problem

#### Droping strongly correlated columns.

```python
df = df.drop(['radius_mean'
            , 'area_mean'
            , 'radius_se'
            , 'area_se'
            , 'perimeter_worst'
            , 'radius_worst'
            , 'area_worst'
            , 'texture_mean'
            , 'concave points_mean'
            , 'concave points_worst'], axis=1)
```

#### Re-plotting co-relation matrix after dropping the strongly corelated variables.

```python
df.corr(method="pearson")
```

### Why dropping Strongly correlated variables from the dataset?

We need to drop strongly correlated varaibles, because they do not convey extra information and won't contribute in prediction but will only increase complexity of the model and reduce the preformance and might cause overfitting of curve while training the model, such models will be more sensitive towards the data, hence they are less reliable. Removing such variables is part of dimension reduction.
